## 用相关性热图（Correlation Heatmap）讲故事

【案例】：如何解读相关热图，苏格兰威士忌的消费大调查。

相关热力图（Coorelation Heatmap）是多变量图（Pair Plot）的一个近亲，可以同时表示很多变量和每一对变量之间的关系。

在多变量图中我们会看到一些表示一对变量之间相关性的数字，或者叫相关性的得分，但如果变量增多，繁杂的数字仍然不利于我们观察。

与其用大量的数字来表示相关性，不如用颜色来赋予数字更加直观的含义，更加直观的反应数值之中蕴含的洞察。

相关热力图的一个很好的应用场景就是找到相关的产品。
