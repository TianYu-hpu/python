为折线图（Line Plot）添加⼀个维度  

【案例】为折线图增加一个维度    

与散点图中的点一样，我们可能经常希望能够区分线图中的多条线。  

与散点图中的点一样的是，我们可以更改线的颜色（Color）、透明度（Transparency），或使用多个数据面板。  

与点的两个区别是，我们可以更改线宽（Width）而不是点大小（Size），还可以更改线型（Linetype，实线、虚线或点线等）而不是点形状（Shape）。
