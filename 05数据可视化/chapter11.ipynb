为了确保所有数据点都可同等感知，可以在 HCL 颜色空间中选择定性的、连续的或发散的色阶变化方式。  

也就是说，在颜色（Hue），色度（Chroma），亮度（Luminance）三个维度中，固定两个维度，变化另外一个维度，来实现不同数据间的比较，同时还能保持所有的数据点都有共同的感知力。
