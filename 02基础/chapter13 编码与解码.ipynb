### 编码表
![编码](./picture/13-8.jpg)
计算机开始发明的时候，用来解决数字计算问题。但是后来发现，只用二进制解决问题还是有限。  
怎么办？那就想办法研究个新的编码呗。  
计算机这种东西最早由老外发明，外国人使用的英语只有26个字母，再加上标点、数字和一些符号也不会太多，老外搞出来一个东西叫做ASCII码，这个比二进制高级了一些。  
ASCII码组合出的256种状态，至此一个字节就使用满了。  
后来，计算机来到中国， 就出现了一个大问题。  
ASCII码最多表示256种状态，我们汉字千千万万， 256个远远不够啊。  
我们科学家就重新编写了一张编码表，叫做GB2312（word中有这种编码方式），它使用2个字节16个比特位，来表示绝大多数（65535）个汉字。后来，为了显示更多的汉字，又研究出一套编码，GBK编码。  
美国人有ASCII码，中国人有GB2312与GBK编码，那阿拉伯人呢？欧洲人呢？  
就像下面的乱码一样，这是咋形成的。就是美国人的编码到了我们中国，我们看不懂；中国人的编码到了欧洲，欧洲人看不懂。   
![编码](./picture/13-9.jpg)
那么，为了能够让全世界人民都能畅通的交流。Unicode应运而生。Unicode把所有语言都统一到一套编码里，可以容纳100多万个符号，这样就不会再有乱码问题了。  
好开森，可以畅通无阻的沟通了。但是由于把所有语言都统一到一套编码里中，就造成Unicode码越来越庞大。比如英文A，之前可以用00010001表示，但是到了Unicode码中，必须使用00000000 00010001来表示。  
太长了。太麻烦了。  
于是，科学家们又针对Unicode码研究出了一种可变长度字符编码，叫做UTF-8（8-bit Unicode Transformation Format），它可以根据不同的符号来变化字节长度。  
我们将编码浓缩到一个图中。 
![编码](./picture/13-10.jpg)
学会了编码表，我们就来看一看计算机中的编码与解码。  
编码encode()与解码decode()  
![编码](./picture/13-11.jpg)
编码encode就是把人类语言转成计算机语言。  
解码decode就是把计算机语言转成人类语言。  
用法如下：  
![编码](./picture/13-12.jpg)

````python
print('但行好事莫问前程'.encode('gbk'))
print('但行好事莫问前程'.encode('utf-8'))
print(b'\xb5\xab\xd0\xd0\xba\xc3\xca\xc2\xc4\xaa\xce\xca\xc7\xb0\xb3\xcc'.decode('gbk'))
print(b'\xe4\xbd\x86\xe8\xa1\x8c\xe5\xa5\xbd\xe4\xba\x8b\xe8\x8e\xab\xe9\x97\xae\xe5\x89\x8d\xe7\xa8\x8b'.decode('utf-8'))

print(type('但行好事莫问前程'))
print(type(b'\xb5\xab\xd0\xd0\xba\xc3\xca\xc2\xc4\xaa\xce\xca\xc7\xb0\xb3\xcc'))
print(type(b'\xe4\xbd\x86\xe8\xa1\x8c\xe5\xa5\xbd\xe4\xba\x8b\xe8\x8e\xab\xe9\x97\xae\xe5\x89\x8d\xe7\xa8\x8b'))
````

这里的计算机语言有一个相同之处，就是最前面都有一个字母‘b’，这是代表它是bytes(字节)类型的数据。  
计算机的编码，就是把字符串类型的数据，转换成bytes类型的数据。  
计算机的解码，就是把bytes类型的数据，转换成字符串类型的数据。  
江湖秘籍：编码时用什么样的编码表，解码的时候就用什么样的编码表。

# 读取文件
![文件读写](./picture/13-13.jpg)
文件读取很简单，就三步。  
![文件读写](./picture/13-14.jpg)
第一步，在编程框的text.txt文件下，随便写点文字内容就可以， 写的“愿你出走半生归来仍是少年“  
![文件读写](./picture/13-15.jpg)
第二步，在编写之后我们在左边的readfile.py写代码。  
![文件读写](./picture/13-16.jpg)
代码如何写？  
首先，使用open()函数打开文件。  
代码片段  
```
myfile = open(r'test.txt','r')
```
![文件读写](./picture/13-17.jpg)
myfile是变量，存放读取的文件。  
第一个r是固定符号。  
'test.txt'是需要读取文件的文件名,在这里, readfile.py文件与test.txt文件需要放在同一文件夹内。  
最后一个‘r’是文件打开模式，‘r’代表read, 意为文件读取。  
现在已经打开了文件。  
第二步，我们要来读取文件中的内容。  
代码片段  
```
myfilecontent = myfile.read()
```
这一句代码，就是使用read()函数，读取myfile变量中文件的数据，将数据放在myfilecontent变量中。  
之后使用print()函数，看看读取到什么内容。  
代码片段  
```
print(myfilecontent) 
```
最后一步，关闭文件。  
代码片段  
```
myfile.close()
```
请运行一下下面代码, 观察终端里的效果。  

````python
myfile = open(r'test.txt','r', encoding='utf-8')
myfilecontent = myfile.read()
print(myfilecontent)
myfile.close()
````

如果我们想写入的数据不是文本，而是图片怎么办？  
![](./picture/13-21.jpg)
我们可以看到里面还有‘wb’模式， 它的意思是以二进制的方式打开一个文件用于写入。  
图片、音频都是以二进制形式存在，所以使用wb模式就好了。后期我们还会遇到。  
顺便再说一下，好多粗心的小伙伴，在打开文件，读写结束后，忘记关闭文件。但是不关闭文件就会占用电脑内存，使得电脑越来越慢，怎么解决呢？  
我们可以使用with方法  
用法如下：  

````python
'''
我们可以看到里面还有‘wb’模式， 它的意思是以二进制的方式打开一个文件用于写入。
图片、音频都是以二进制形式存在，所以使用wb模式就好了。后期我们还会遇到。
顺便再说一下，好多粗心的小伙伴，在打开文件，读写结束后，忘记关闭文件。
但是不关闭文件就会占用电脑内存，使得电脑越来越慢，怎么解决呢？
我们可以使用with方法
'''
with open(r'test.txt','a',encoding='utf-8') as myfile:
    myfile.write('但行好事莫问前程\r\n')
````
